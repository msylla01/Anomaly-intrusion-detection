```python
import sqlite3
cnx = sqlite3.connect(':memory:')
```

```python
data.to_sql('arreste', con=cnx, if_exists='append', index=False)
```

```python
def sql_query(query):
    return pd.read_sql(query, cnx)
```

```python
query = '''
SELECT COUNT(area),area,year,Age
FROM arreste
WHERE year='2018'
GROUP BY area;
'''
sql_query(query)
```

```python
query = '''
SELECT area,year,Age
FROM arreste
WHERE year='2018';
'''
d1=sql_query(query)
```
